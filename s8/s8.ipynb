```python
import numpy as np
import numpy.linalg as la
from math import acos, sin, cos

R = 6370 # km

lr = np.deg2rad(-22) # lengdarbaugur Reykjavík
br = np.deg2rad(64) # breiddarbaugur Reykjavík
rvk = np.array([R*sin(lr)*cos(br), R*cos(lr)*cos(br), R*sin(br)])

ll = np.deg2rad(0) # lengdarbaugur London
bl = np.deg2rad(51.5) # breiddarbaugur London
lon = np.array([R*sin(ll)*cos(bl), R*cos(ll)*cos(bl), R*sin(bl)])

def hnattstada(x,y):
    theta = np.degrees(acos((x@y)/(la.norm(x)*la.norm(y))))
    return theta

print(hnattstada(rvk,lon))
```

```python
def lengd(x,y,R):
    L = R * np.deg2rad(hnattstada(x,y))
    return L

print(lengd(rvk,lon,R))
```
